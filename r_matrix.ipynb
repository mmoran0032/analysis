*R*-Matrix
========

From the AZURE calculation, we can view and analyze the predicted *R*-matrix results and compare it to our two "isolated" resonances. I say isolated only because our lower resonance does have a close, lower cross section, resonance just above it. I'm not 100% sure how exactly to do this, but I'll figure it out.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

import pyne
import sap
pyne.__version__, sap.__version__
```

```python
r_matrix = '../azure_calc.tsv'
energy, *_, dxs, _ = np.loadtxt(r_matrix, unpack=True)

plt.semilogy(energy, dxs, 'k-')
plt.xlabel('Center of Mass Energy [MeV]')
plt.ylabel('Differential Cross Section [sr/b]')
plt.ylim(1e-9, 1e-1)
plt.xlim(0.8, 1.4)
plt.show()
```

```python
m_p = 1.00782503
m_alpha = 4.00260325
m_mg24 = 23.98504170
m_al27 = 26.98153863

e_lab = energy * (m_p + m_al27) / m_al27
```

```python
plt.semilogy(e_lab, dxs, 'k-')
plt.xlabel('Lab Energy [MeV]')
plt.ylabel('Differential Cross Section [sr/b]')
plt.ylim(1e-9, 1e-1)
plt.xlim(0.8, 1.4)
plt.show()
```

```python
f = np.load('../extracted_data/cross_sections.npz')
e_p, xs = f['e_p'], f['xs']
```

```python
plt.semilogy(e_lab, dxs, 'k-', linewidth=1.5, alpha=0.5)
plt.semilogy(e_p, xs, linestyle='none', marker='.', markersize=8)

plt.xlabel('Lab Energy [MeV]')
plt.ylabel('Differential Cross Section [sr/b]')
plt.ylim(1e-7, 1e-1)
plt.xlim(1.1, 1.45)
plt.show()
```
