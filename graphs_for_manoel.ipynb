```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy
```

```python
data = numpy.array([0, 46, 45, 298, 6293, 31649, 58027, 50590, 39051,
                    31528, 23413, 15539, 8303, 3630, 1183, 239, 34, 0])
strips = numpy.arange(-1, 17, 1)

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

fig = plt.figure(figsize=(6, 6))
axis = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis.set_title(r'$E_p = 1.36$~MeV Resonance', fontdict={'fontsize': 14})
axis.set_xlabel(r'Detector Strip')
axis.set_ylabel(r'Counts')
axis.plot(strips, data, drawstyle='steps-mid')
axis.set_xlim((-0.5, 15.5))
plt.xticks(numpy.arange(0, 16, 1))

fig.savefig('high_resonance_linear.png')

plt.show()
```
